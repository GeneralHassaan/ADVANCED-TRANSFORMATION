```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
file_1 = 'Data_Sets/s_info.csv'
df_char = pd.read_csv(file_1)
df_char.head()
```

```python
df_char[['hero_names','Publisher']] = df_char['Hero|Publisher'].str.split('|',expand=True)
df_char.head(4)
```

```python
df_char.drop(columns = ['Hero|Publisher'], inplace = True)
```

```python
df_char.head(5)
```

```python
df_char[['Height','Weight']] = df_char['Measurements'].str.split(',',expand=True)
df_char.head(2)
```

```python
#df_char.drop(columns = ['Hero_names', 'Measurements'], inplace = True)
```

```python
to_replace = ['{','cm']

for char in to_replace:
    df_char['Height'] = df_char['Height'].str.replace(char,'', regex=False)
    
df_char['Height'].head()
```

```python
to_replace = ['Height',':',"'"]

for char in to_replace:
    df_char['Height'] = df_char['Height'].str.replace(char,'', regex=False)
    
df_char['Height'].head()
```

```python
to_replace = ['kg']

for char in to_replace:
    df_char['Weight'] = df_char['Weight'].str.replace(char, '', regex=False)
    
df_char['Weight'].head()
```

```python
to_replace = ['Weight', ':' , "'", "}", ".0"]

for char in to_replace:
    df_char['Weight'] = df_char['Weight'].str.replace(char, '', regex=False)
    
df_char['Weight'].head()
```

```python
df_char["Weight"] = pd.to_numeric(df_char["Weight"])
df_char["Height"] = pd.to_numeric(df_char['Height'])
```

```python
df_char.head(3)
```

```python
power_set = pd.read_csv('Data_Sets/s_powers.csv')
power_set.head(5)
```

```python
df = pd.merge(df_char, power_set, how='inner', on = 'hero_names')
df.head(4)
```

```python
df['Powers'].head()
```

```python
df_powers = df['Powers'].str.split(",")
df_powers.head(4)
```

```python
exploded = df_powers.explode('Powers')
```

```python
cols_to_make = exploded.dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df[col] = df["Powers"].str.contains(col)
df.head()
```
